## OD Tech

```python
def odiqr(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3 - q1
    low = q1 - (1.5 * iqr)
    high = q3 + (1.5 * iqr)
    m = df.mean()
    df = df.apply(lambda x : m if x < low else (m if x > high else x ) )
    return df
```

```python
def odmsd(df):
    m = round(df.mean(),2)
    s = round(df.std(),2)
    low = round(m-(3*s),2)
    high = round(m+(3*s),2)
    ft1 = df[df<low]
    ft2 = df[df>high]
    df = df.map(lambda x : low if x < low else (high if x > high else x ))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold , cross_val_score
from sklearn.ensemble import RandomForestClassifier 
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=.20)
```

```python
kf = KFold(n_splits=12)
```

```python
rfc = RandomForestClassifier()
```

```python
algo = [rfc]
```

### Without Feature Scaling

```python
for i in algo:
    i.fit(Xtrain,ytrain)
    s = i.score(Xtest,ytest)
    print(f'{i} = {s}')
```

### With Feature Scaling

```python
ss = StandardScaler()
```

```python
ss.fit(Xtrain)
```

```python
Xtrain_ss = ss.transform(Xtrain)
```

```python
Xtest_ss = ss.transform(Xtest)
```

```python
for i in algo:
    i.fit(Xtrain_ss,ytrain)
    s = i.score(Xtest_ss,ytest)
    print(f'{i} = {s}')
```

### With Cross Validation

```python
for i in algo:
    s = cross_val_score(i,X,y,cv = kf)
    print(f'{i} = {s.mean()}')
```
